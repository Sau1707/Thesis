# Fundamentals of returns
- Average return not good to look how an asset behaves
- Average return doesn't mean that the end value is the same


Multi period return (compounded return)

$$R_{t, t+n} = (1 + R_{t, t+1}) \cdot (1 + R_{t+1, t+2}) \cdot \ldots \cdot (1 + R_{t+n-1, t+n}) - 1$$

Annualized return

$$R_{annualized} = (1 + R_{t})^{12} - 1$$

Return on the asset

$$R_{t, t+1} = \frac{P_{t+1} - P_t}{P_t} = \frac{P_{t+1}}{P_t} - 1$$


```python
prices = pd.DataFrame({"A": [8.70, 8.91, 8.71, 8.43, 8.73], "B": [10.66, 11.08, 10.71, 11.59, 12.11]})
returns = prices.pct_change()
prices.plot()
returns.plot.bar()

returns.std()
returns.mean()
```
